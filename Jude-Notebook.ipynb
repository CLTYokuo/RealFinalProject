```python
import matplotlib.pyplot as plt

data = pd.read_csv('Final Block Value - Sheet1.tsv', sep='\t')

data['Median Sales Price'] = pd.to_numeric(data['Median Sales Price'].replace('[\$,]', '', regex=True))

data.plot(kind='bar', x='Neighborhoods', y='Median Sales Price', legend=False)
plt.title('Bar Graph of Final Block Values')
plt.xlabel('Neighborhoods')
plt.ylabel('Median Sales Price')
plt.show()
```

```python
data = pd.read_csv('Final Block Value - Sheet1.tsv', sep='\t')

data['Median Sales Price'] = pd.to_numeric(data['Median Sales Price'].replace('[\$,]', '', regex=True))

data_sorted = data.sort_values(by='Median Sales Price', ascending=False)

top_10_data = data_sorted.head(10)

top_10_data.plot(kind='bar', x='Neighborhoods', y='Median Sales Price', legend=False)
plt.title('Top 10 of Final Block Values')
plt.xlabel('Neighborhoods')
plt.ylabel('Median Sales Price')
plt.show()
```
