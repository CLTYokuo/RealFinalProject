# Main Metric

For our final metric we decided to use a formula to give us the best neighborhood. Our formula used the median household income as the base number and then we multiplied this by the formula numbers, listed in the table, gathered from both the green spaces submetric and median sales price to give us the final number. 

```python
final_df = pd.read_csv('MainMetric - Sheet1 (4).tsv', sep='\t')
final_df['Total Score'] = final_df['Formula Numbers(Green)'] * final_df["Formula Numbers(Sales)"] * final_df["Median HH Income"]

print(final_df)
```

```python
import matplotlib.pyplot as plt

final_df.plot(kind='bar', x='Neighborhoods', y='Total Score', legend=False)
plt.title('Bar Graph of Total Scores')
plt.xlabel('Neighborhoods')
plt.ylabel('Score')
plt.show()
```

```python
top_10_final_df = final_df.head(10)

top_10_final_df.plot(kind='bar', x='Neighborhoods', y='Total Score', legend=False)
plt.title('Top 10 of Total Scores')
plt.xlabel('Neighborhoods')
plt.ylabel('Total Score')
plt.show()
```

```python
low_10_final_df = final_df.tail(10)

low_10_final_df.plot(kind='bar', x='Neighborhoods', y='Total Score', legend=False)
plt.title('Bottom 10 of Total Scores')
plt.xlabel('Neighborhoods')
plt.ylabel('Total Score')
plt.show()
```
