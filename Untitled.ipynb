Here is a bar graph representing the data based on Median HH income

```python
import matplotlib.pyplot as plt
fp = 'DataSetReal.tsv'
df = pd.read_csv(fp, delimiter = '\t')
dfSorted= df.sort_values(by='median HH income')

#create bar graph 
plt.bar(dfSorted['Neighborhood'],
dfSorted['median HH income'])

plt.xlabel('Neighborhood')
plt.ylabel('Median HouseHold Income')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```
